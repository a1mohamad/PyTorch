# Interactive CartPole with OpenAI Gym

This notebook demonstrates the initial setup and interaction with the **CartPole-v1** environment using the `gym` library and `matplotlib` for visualization.

---

## üõ†Ô∏è 1. Environment Initialization
The notebook begins by importing the necessary libraries and creating the simulation environment.

* **Libraries:** Uses `gym` for the reinforcement learning environment and `matplotlib.pyplot` for rendering the visual state.
* **Initialization:** The environment is created using `gym.make('CartPole-v1', render_mode='rgb_array')`, which allows the state to be captured as an image array.

---

## üìä 2. Space Definitions
The notebook inspects the constraints of the CartPole problem:

* **Observation Space:** A `Box` (continuous) space with 4 variables representing:
    1. Cart Position
    2. Cart Velocity
    3. Pole Angle
    4. Pole Angular Velocity
* **Action Space:** A `Discrete(2)` space, meaning the agent has exactly two possible choices:
    * **0**: Push the cart to the left.
    * **1**: Push the cart to the right.

---

## üîÑ 3. Stepping Through the Environment
The notebook demonstrates the basic control loop of a reinforcement learning task:

* **Resetting:** `env.reset()` initializes the environment to a starting state and returns the first observation.
* **Action Execution:** * `env.step(action=0)`: Applies a force to the left.
    * `env.step(action=1)`: Applies a force to the right.
* **Step Output:** Every step returns a tuple containing the **next observation**, the **reward** (1.0 for staying upright), and flags indicating if the episode is finished.

---

## üñºÔ∏è 4. Rendering the State
To visualize the cart and pole, the notebook captures the current state of the environment and displays it:

* **Capture:** `env.render()` generates an RGB image array of the current simulation frame.
* **Visualization:** `plt.imshow(img)` is used to display the cart and the balanced pole within the notebook.

```python
import gym
import matplotlib.pyplot as plt
```

```python
env = gym.make('CartPole-v1', render_mode= 'rgb_array')
```

```python
env.observation_space
```

```python
env.action_space
```

```python
env.reset()
```

```python
env.step(action= 0)
```

```python
env.step(action= 1)
```

```python
img = env.render()
```

```python
plt.imshow(img)
plt.show()
```
